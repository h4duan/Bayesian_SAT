```python
#True hyperparameters
theta = 0.78
c1 = 0.23
c2 = 0.65
d1 = 1 - c1
d2 = 1 - c2
```

```python
num_iterations = 500
```

```python
import numpy as np
```

```python
coefficients = np.zeros(1)
first_moment = np.zeros(num_iterations+1)
```

```python
first_moment[0] = 0.5
coefficients[0] = 1
```

```python
true_p = theta * c1 + (1 - theta) * c2
```

```python
def moment(coefficient):
    M1 = 0
    for i in range(len(coefficient)):
        M1 += coefficient[i] / (i+2)
    return M1
```

```python
def probability(coefficient):
    p = 0
    for i in range(len(coefficient)):
        p += coefficient[i] / (i+1)
    return p
```

```python
import random
```

```python
epsilon = 0.000001
```

```python
for i in range(num_iterations):
    print(probability(coefficients))
    print(coefficients)
    r = random.uniform(0, 1)
    if(r < true_p):
        denominator = (c1 - c2) * first_moment[i] + c2
        coefficients = np.append(coefficients, coefficients[i] * (c1-c2))
        for j in range(i, 0, -1):
            coefficients[j] = coefficients[j-1] * (c1-c2) + c2*coefficients[j]
        coefficients[0] = c2*coefficients[0]
        coefficients = coefficients / probability(coefficients)
    else:
        denominator = (d1 - d2) * first_moment[i] + d2
        coefficients = np.append(coefficients, coefficients[i] * (d1-d2))
        for j in range(i, 0, -1):
            coefficients[j] = coefficients[j-1] * (d1-d2) + d2*coefficients[j]
        coefficients[0] = d2*coefficients[0]
        coefficients = coefficients / probability(coefficients)
    first_moment[i+1] = moment(coefficients)      
            
```
